```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
%matplotlib nbagg

N = 400
a = 4
array = np.linspace(0.1,30,N)
f1 = np.zeros((N,6))


for i in range(N):
    u = array[i]
    def func(x):
        return x+a*(2*x**2+3*x**3+4*x**4+5*x**5+6*x**6)-array[i]
    x0 = fsolve(func, 1)
    f1[i,0] = x0/array[i]
    
    
for i in range(2,7):
    f1[:,i-1] = i*a*array**(i-1)*f1[:,0]**i

total = f1[:,0]+f1[:,1]+f1[:,2]+f1[:,3]+f1[:,4]+f1[:,5]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid(color='k', linestyle='--', linewidth=.2)
plt.suptitle('Concentration Fractions')
ax.set_xlabel('u',fontsize=15)
ax.set_ylabel('f',fontsize=15)

plt.plot(array,total,label="sum of p's")
for i in range(0,6):
    string = 'p' + str(i+1)
    plt.plot(array,f1[:,i],label=string)
    
plt.legend(bbox_to_anchor=(.75, .99), loc=2, borderaxespad=0.)
plt.show()

plt.savefig('concentration_fractions2.pdf')

```
